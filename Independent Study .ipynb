```python
import pandas as pd
import plotly.graph_objects as go

def create_bar_and_scatter(file_path, par1,par2, par3):
    df = pd.read_csv(file_path)
    fig = go.Figure()
    fig.add_trace(go.Bar(x=df[par1], y=df[par2], name='Bar Plot'))
    fig.add_trace(go.Scatter(x=df[par1], y=df[par3], mode='markers', name='Scatter Plot'))
    fig.update_traces(showlegend=True)
    fig.update_layout(legend_traceorder="reversed")
    fig.show()
create_bar_and_scatter("https://raw.githubusercontent.com/fivethirtyeight/""data/master/college-majors/recent-grads.csv", 'Men', 'Women', 'Full_time')
```

```python
import plotly.graph_objects as go
import pandas as pd

def create_parcoords_plot(file_path, list_of_param, color, max_min):
    df = pd.read_csv(file_path)
    
    dimensions = []
    for par in list_of_param:
        dimension = dict(label=par, values=df[par])
        dimensions.append(dimension)
    
    fig = go.Figure(data=go.Parcoords(
        line=dict(color=color,
                  colorscale='Rainbow',
                  showscale=True,
                  cmax=max_min.max(),
                  cmin=max_min.min()),
        dimensions=dimensions
    ))
    
    fig.show()
df = pd.read_csv("C:/Users/YoYo Samon/Downloads/all_Aug_2019.csv")
color = df['runtime']
max_min = df['runtime']
list_of_param = ['power_cap', 'thread_count', 'task_count', 'year', 'bw_level', 'runtime']
create_parcoords_plot("C:/Users/YoYo Samon/Downloads/all_Aug_2019.csv", list_of_param, color, max_min)
```
